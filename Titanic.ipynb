```
#All this to calculate average age (and some accurate population)
totAge1 = 0
count1 = 0
valid1 = 1
totAge2=0
count2 = 0
valid2 = 1
totAge3 =0
count3=0
valid3 = 1
for i in range(1,418):
  if testdataset.at[i,"Pclass"] == 1:
    count1+=1
    try:
      totAge1+=int(testdataset.at[i,"Age"])
      valid1+=1
    except ValueError:
      totAge1+=0
  elif testdataset.at[i,"Pclass"] == 2:
    count2+=1
    try:
      totAge2+=int(testdataset.at[i,"Age"])
      valid2+=1
    except ValueError:
      totAge2+=0
  elif testdataset.at[i,"Pclass"] == 3:
    count3+=1
    try:
      totAge3+=int(testdataset.at[i,"Age"])
      valid3+=1
    except ValueError:
      totAge3+=0
for i in range(1,891):
  if traindataset.at[i,"Pclass"] == 1:
    count1+=1
    try:
      totAge1+=int(traindataset.at[i,"Age"])
      valid1+=1
    except ValueError:
      totAge1+=0
  elif traindataset.at[i,"Pclass"] == 2:
    count2+=1
    try:
      totAge2+=int(traindataset.at[i,"Age"])
      valid2+=1
    except ValueError:
      totAge2+=0
  elif traindataset.at[i,"Pclass"] == 3:
    count3+=1
    try:
      totAge3+=int(traindataset.at[i,"Age"])
      valid3+=1
    except ValueError:
      totAge3+=0
mean1=totAge1/valid1
mean2=totAge2/valid2
mean3=totAge3/valid3
print(mean1,mean2,mean3)
```

```
#variance and standard deviation stuff to generate pseudo random ages for missing passengers based off cabin class
import math
var1 = 0
var2 = 0
var3 = 0
for i in range(1,418):
  if testdataset.at[i,"Pclass"] == 1:
    try:
      var1+=(int(testdataset.at[i,"Age"])-mean1)**2
    except ValueError:
      var1+=0
  elif testdataset.at[i,"Pclass"] == 2:
    try:
      var2+=(int(testdataset.at[i,"Age"])-mean2)**2
    except ValueError:
      var2+=0
  elif testdataset.at[i,"Pclass"] == 3:
    try:
      var3+=(int(testdataset.at[i,"Age"])-mean3)**2
    except ValueError:
      var3+=0
for i in range(1,891):
  if traindataset.at[i,"Pclass"] == 1:
    try:
      var1+=(int(traindataset.at[i,"Age"])-mean1)**2
    except ValueError:
      var1+=0
  elif traindataset.at[i,"Pclass"] == 2:
    try:
      var2+=(int(traindataset.at[i,"Age"])-mean2)**2
    except ValueError:
      var2+=0
  elif traindataset.at[i,"Pclass"] == 3:
    try:
      var3+=(int(traindataset.at[i,"Age"])-mean3)**2
    except ValueError:
      var3+=0
var1*=(1/valid1)
var2*=(1/valid2)
var3*=(1/valid3)
var1 = math.sqrt(var1)
var2 = math.sqrt(var2)
var3 = math.sqrt(var3)
print(var1,var2,var3)
```

```
import numpy as np
from numpy import random
for missingAge in range(1,891):
  if traindataset.at[missingAge,"Pclass"] == 1:
    try:
      int(traindataset.at[missingAge,"Age"])
    except ValueError:
      traindataset.at[missingAge,"Age"] =abs(int(np.random.normal(loc=mean1, scale=var1)))
  if traindataset.at[missingAge,"Pclass"] == 2:
    try:
      int(traindataset.at[missingAge,"Age"])
    except ValueError:
      traindataset.at[missingAge,"Age"] =abs(int(np.random.normal(loc=mean2, scale=var2)))
  if traindataset.at[missingAge,"Pclass"] == 3:
    try:
      int(traindataset.at[missingAge,"Age"])
    except ValueError:
      traindataset.at[missingAge,"Age"] =abs(int(np.random.normal(loc=mean3, scale=var3)))
for missingAge in range(1,418):
  if testdataset.at[missingAge,"Pclass"] == 1:
    try:
      int(testdataset.at[missingAge,"Age"])
    except ValueError:
      testdataset.at[missingAge,"Age"] =abs(int(np.random.normal(loc=mean1, scale=var1)))
  if testdataset.at[missingAge,"Pclass"] == 2:
    try:
      int(testdataset.at[missingAge,"Age"])
    except ValueError:
      testdataset.at[missingAge,"Age"] =abs(int(np.random.normal(loc=mean2, scale=var2)))
  if testdataset.at[missingAge,"Pclass"] == 3:
    try:
      int(testdataset.at[missingAge,"Age"])
    except ValueError:
      testdataset.at[missingAge,"Age"] =abs(int(np.random.normal(loc=mean3, scale=var3)))
```

```
import pandas as pd


df = pd.read_csv("/content/Titanic/csv/train.csv")


#for ix in range(len(df)):
  #print(df.loc[[ix], "Name"])

pClass = {1:1, 2:2, 3:3}
gender = {"male": 1, "female": 2}

df["Sex"] = df["Sex"].map(gender)
df["Pclass"] = df["Pclass"].map(pClass)
df["Name"] = df["Name"].astype("category").cat.codes
df["Ticket"] = df["Ticket"].astype("category").cat.codes
df["Fare"] = df["Fare"].astype("category").cat.codes
#df["Cabin"] = df["Cabin"].astype("category").cat.codes
df["Cabin"] = 0
df["Embarked"] = df["Embarked"].astype("category").cat.codes
df.head(100)
```

```
import sklearn
from sklearn import svm, preprocessing

x = df.drop("Survived", axis=1).values
x = preprocessing.scale(x)
y = df["Survived"].values

test_size = 200

x_train = x[:-test_size]
y_train = y[:-test_size]

x_test = x[-test_size:]
y_test = y[-test_size:]


clf = svm.SVR(kernel="linear")
clf.fit(x_train, y_train)
```

```
clf.score(x_test, y_test)
```
